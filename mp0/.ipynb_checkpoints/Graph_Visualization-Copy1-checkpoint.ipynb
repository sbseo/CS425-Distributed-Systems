```python
"""
node1.txt: Data Structure

nodeTime nodeData loggerTime bytesTransfered

"""

NUM_OF_NODE = 3

"""
Create Lists :
        F1, F2, F3, F4, F5, F6, F7, F8
        DELAYS[0], DELAYS[1], DELAYS[2], ...

"""
FILES, DELAYS = list(), list()
for i in range(1, NUM_OF_NODE+1):
    F = open("node" + str(i) +".txt","r")
    FILES.append(F)
    delayList = list()
    DELAYS.append(delayList)
    
# F1, F2, F3 = open("node1.txt","r"), open("node2.txt","r") , open("node3.txt","r") 
# delay_n1, delay_n2, delay_n3= list(), list(), list()
"""
Get Delay, Local Server Time(1min, 2mins)
"""
for i, file in enumerate(FILES):
    for line in file:
        sen = line.split() 
        # Code from here
        delay = float(sen[2]) - float(sen[0])
        DELAYS[i].append(delay)
        
```

```python
cutNum = min([len(DELAYS[i]) for i in range(NUM_OF_NODE)])  # Choose Shortest Timestamps
print(cutNum)
```

```python
df1 = pd.DataFrame(DELAYS).T
df1 = df1.rename(columns={0: "Node1", 1: "Node2", 2: "Node3"})
df1 = df1[:cutNum]
df1
# df1.to_csv(r'./log.csv') # Save to CSV file
```
