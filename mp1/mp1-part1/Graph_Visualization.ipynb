### Graph Visualization (All scenarios)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Extremely important
sns.set(style="whitegrid", font_scale=1.35, rc={'figure.figsize':(11.7,5.27)}) #Change Font Size 

# Configuration
SCENARIO_NUM = 1
NUM_OF_NODE = 3
TIME = 100
```

### Data Structure of Bandwidth CSV

node1_bw    node2_bw    node3_bw 


```python
bw = pd.read_csv("results/s" + str(SCENARIO_NUM) + "_bandwidth.csv")
bw = bw[:TIME].T
bw
```

### Data Structure of Delay CSV

Delay

```python
d = pd.read_csv("results/s" + str(SCENARIO_NUM) + "_delay.csv").T
d
```

```python
# Setting
numTransactions = range(0, d.shape[1]) # Delay
timeRange = range(0, TIME) # Bandwidth

plt.subplots(figsize=(16, 10)) 
plt.subplots_adjust(hspace=.45)

# Delay
ax1= plt.subplot(211)
plt.plot(numTransactions, d.T, marker='h')

# plt.plot(timeRange, s1d_maxVal, marker='s')
# plt.plot(timeRange, s1d_median, marker='*')
# plt.plot(timeRange, s1d_quantile90, marker='^')
# ax1.legend(['Min', 'Max', 'Median', '90th Percentile'])
ax1.set(xlabel='Transactions', ylabel='delay(millisecs)', title='Scenario ' + str(SCENARIO_NUM) + ' Delay', label="max")
plt.grid(b=None, which='major', axis='x')

# Bandwidth
ax2 = plt.subplot(212)
plt.plot(timeRange, bw.T, marker='h')

nodeList = list()
for i in range(0, NUM_OF_NODE):
    s = str("Node") + str(i+1)
    nodeList.append(s)
ax2.legend(nodeList)
ax2.set(xlabel='time(sec)', ylabel='bandwidth(byte/sec)', title='Scenario ' + str(SCENARIO_NUM) + ' Delay', label="max")
plt.grid(b=None, which='major', axis='x')
```
